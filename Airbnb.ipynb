# These are column names in a data table and represent the following information:

 1. 'id': a unique identifier for each listing
 2. 'name': the name of the listing
 3. 'host_id': the unique identifier for the host of the listing
 4. 'host_name': the name of the host of the listing
 5. 'neighbourhood_group': a grouping of neighborhoods
 6. 'neighbourhood': the neighborhood where the listing is located
 7. 'latitude': the latitude coordinate of the location of the listing
 8. 'longitude': the longitude coordinate of the location of the listing
 9. 'room_type': the type of room being offered (e.g. private room, shared room, entire home/apartment)
 10. 'price': the price per night for the listing
 11. 'minimum_nights': the minimum number of nights a guest is required to stay at the listing
 12. 'number_of_reviews': the total number of reviews received for the listing
 13. 'last_review': the date of the most recent review for the listing
 14. 'reviews_per_month': the average number of reviews per month for the listing
 16. 'calculated_host_listings_count': the number of listings a host has
 17. 'availability_365': the number of days in a year when the listing is available for booking
 18. 'number_of_reviews_ltm': the number of reviews received in the last 12 months
 19. 'license': the license number for the listing (if applicable)
